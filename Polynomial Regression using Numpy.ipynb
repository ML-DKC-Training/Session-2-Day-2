```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Creating the data set for polynomial regression
x = np.linspace(-2, 2, num=100)
y = 6*x**2 + 5*x + np.random.normal(loc=.3,scale=.90,size=len(x))
plt.plot(zip(y,x))
plt.show()

print x.shape
```

```python
# Splitting the dataset into test and train

split_ratio = 0.8

idxs = np.arange(len(x))
np.random.shuffle(idxs)

split_idx = int(0.8*len(x))

test_idxs = idxs[split_idx:]
train_idxs = idxs[:split_idx]

X = np.hstack((np.ones(shape=(x.shape[0],1)), x.reshape(-1,1),x.reshape(-1,1)**2))
train_X = X[train_idxs]
test_X = x[test_idxs]
train_y = y.reshape(-1,1)[train_idxs]
test_y = y.reshape(-1,1)[test_idxs]

print train_X.shape, train_y.shape
```

```python
def get_dW(train_X, train_y, W):

    # y_pred calculation
    y_pred = np.matmul(train_X, W)

    # Weight update
    yTx = np.matmul(np.transpose(train_y), train_X)
    xTx = np.matmul(np.transpose(train_X),train_X)
    wTxTx = np.matmul(np.transpose(W), xTx)
    dW = (-2*yTx + 2*wTxTx)*1./train_y.shape[0]
    
    return np.transpose(dW)
 
def get_loss(train_X, train_y, W):

    # y_pred
    y_pred = np.matmul(train_X, W)

    # Error
    e = train_y - y_pred

    #Loss computation
    loss = np.matmul(np.transpose(e), e)/4000.

    print 'loss:', np.asscalar(loss)
```

```python
#Learning rate
lr = 1e-2

# Initilize weights
W = np.random.normal(loc=0, scale=1e-3, size=(train_X.shape[1],1))

for i in range(2000):
    dW = get_dW(train_X, train_y, W)
    W = W - lr*dW
    get_loss(train_X, train_y, W)
#     print i, W
    
print W
```
